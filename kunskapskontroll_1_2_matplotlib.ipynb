### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

**ANSWER:** The inline method as is done in Python, or the object oriented method

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

**ANSWER:** Figure is the whole object. Axes are the subparts of the figure each one a separate area of it with their own lines and properties. Axis is part of an axes containing a line and its belonging details such as labels, markers and any other primitives. An artist is any detail on any level in an illustration, from figure to axes, and axis, and any other primitives.

### When plotting in Matplotlib, what is the expected input data type?

**ANSWER:** Numpy arrays

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 300)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
# we use np.abs(x) as this one may result in complex values
# that cannot be plotted
# ax.plot(x, x**np.cos(x**3)*np.exp(np.sin(x)))
ax.plot(x, np.abs(x)**np.cos(x**3)*np.exp(np.sin(x)))
ax.set_title(r'Plotting $x^{cos(x**3)} \times e^{sin(x)}$')
ax.set_xlabel(r'input $[\mu N]$')
ax.set_ylabel('output $[V]$')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# MY CODE
# define a figure
fig, axs = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True)
fig.suptitle('Test of Plots')

min_x = random_data_x.min()
max_x = random_data_x.max()
min_y = random_data_y.min()
max_y = random_data_y.max()
margin = 0.5

# define the scatter plot
axs[0].scatter(random_data_x, random_data_y, c=random_data_x + max_y, s=10*(random_data_y + max_y), alpha=0.6, edgecolors='black')
axs[0].set_xlim(min_x-margin, max_x+margin)
axs[0].set_ylim(min_y-margin, max_y+margin)
print(f"min(x)={min_x}, max(y)={max_x}, min(y)={min_y}, max(y)={max_y}")
axs[0].set_xlabel('Input')
axs[0].set_ylabel('Output')
axs[0].set_title('Random Data')

#define the bar plot
axs[1].bar(names, values)
axs[1].set_xlabel('Type')
axs[1].set_ylabel('Number')
axs[1].set_title('Fuit Inventory')
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
